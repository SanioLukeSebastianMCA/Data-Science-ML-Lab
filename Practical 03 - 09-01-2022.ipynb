```python
import pandas as pd
import numpy as np
```

```python
# 1.1.	How to create Series with nd array
arr=np.array([10,15,18,22])
s = pd.Series(arr)
print(s)
```

```python
# 1.2.	How to create Series with Mutable index
arr=np.array(['a','b','c','d'])
s=pd.Series(arr,
index=['first','second','third','fourth'])
print(s)
```

```python
# 1.3.	Creating a series from a Dictionary
d= {'Name':'Sanio Luke', 'Roll No':'35', 'Designation':'Android Developer'}
s= pd.Series(d)
print(s)
```

```python
# 1.4.	Print all the values of the Series by multiplying them by 2.
s= pd.Series([1,2,3,4,5])
print(s*2)
```

```python
# 1.5.	Print Square of all the values of the series.
s= pd.Series([1,2,3,4,5])
print(s**2)
```

```python
# 1.6.	Print all the values of the Series that are greater than2
s= pd.Series([1,2,3,4,5])
print(s[s>=4])
```

```python
# 1.7.	Addition of two series
s1= pd.Series([1,2,3,4,5],['a','b','c','d','e'])
s2= pd.Series([49,48,47,46,45],['a','b','c','d','e'])
print(s1+s2)
```

```python
# 1.8.	Print the first and last 5 elements of a series
arr= pd.array([10,45,53,81,2,23,86,24,44,21])
s= pd.Series(arr)
print(f"The array is - {arr}\n")
print("First 5 Numbers in the given array are - ")
print(s.head(5))

print("Last 5 Numbers in the given array are - ")
print(s.tail(5))
```

```python
# 1.9.	Print the values from index 0 to 5
arr= np.array([10,15,20,25,30,75,35,57,22,11])
s= pd.Series(arr)
print('The given array is ')
print(s)
print("The values from 0 to 5 are -")
print(s[:5])
```

```python
# 1.10.	Selection Using loc, iloc index label
arr= np.array([10,15,20,25,30])
s= pd.Series(arr)
print('The given array is ')
print(s)
print(s.iloc[:2])
print(s.iloc[2:3])
```

```python
# 1.11.	Retrieve subsets of data using slicing
arr= pd.Series([10,15,20,25,30],['A','B','C','D','E'])
s= pd.Series(arr)
print('The given array is ')
print(s)

print(s[1:5:2])
```

```python
# 1.12. create Dataframe From Series
print(pd.DataFrame(pd.Series([1,2,3,4,5,6,7,8,9])))
```

```python
# 1.13. DataFrame from List of Dictionaries
print(pd.DataFrame({'name':[2,4,5,7,8], 'new':[4,6,8,9,1]}))
```

```python
# 1.14. Display the first 5 rows of data frame
print(pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1]}).head())
```

```python
# 1.15. Select the last two columns of the data frame
print(pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1], 'value':[1,2,3,4,5,6,7,8]}).iloc[: , 1:])
```

```python
# 1.16. Add two data frames
print(pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1]})+ pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1]}))
```

```python
# 1.17. Demonstrate deletion, and renaming of columns
val=pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1], 'value':[1,2,3,4,5,6,7,8], 'value2':[1,2,3,4,5,6,7,8]})
print(val.drop('name',axis=1))
val.rename(columns={'name':'name2'}, inplace=True)
print(val)
```

```python
# 1.18. Demonstrate concat, Merge operations in data frame
print(pd.concat([pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1]}),pd.DataFrame({'name':[2,4,5,7,8,5,2,4], 'new':[4,6,8,9,1,7,9,1]})], ignore_index=True))
```

```python
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 'Age':[27, 24, 22, 32],} 
data2 = {'key': ['K0', 'K1', 'K2', 'K3'], 'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2) 
# print(df, "\n\n", df1) 
print(pd.merge(df, df1, on='key'))
```

```python
# 1.19. Write a Pandas program to join the two given dataframes along rows and assign all data

s1 = pd.DataFrame({'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 'marks': [200, 210, 190, 222, 199]})
s2 = pd.DataFrame({'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 'marks': [201, 200, 198, 219, 201]})
print("Original DataFrames:")
print(s1)
print("-------------------------------------")
print(s2)
print("\ntwo dataframes along rows:")
result_data = pd.concat([s1, s2])
print(result_data)
```

```python
# 1.20. save dataframe to csv file
result_data.to_csv('output.csv')
data=pd.read_csv('output.csv')
print(data)
```
